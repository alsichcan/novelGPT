```python
import os
import time
import requests
from bs4 import BeautifulSoup

error = set()

def login(username, password):
    login_url = 'https://ssl.syosetu.com/login/login/'
    data = {
        'narouid': username,
        'pass': password
    }

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:97.0) Gecko/20100101 Firefox/97.0'
    }
    
    session = requests.Session()
    session.headers.update(headers)
    response = session.post(login_url, data=data)

    
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        if "ゼサン [ID:2308154] でログイン中" in soup.text:
            print("로그인 성공!")
            return session
        else:
            print("로그인 에러!")
            return None
    else:
        print(f"HTTP 에러 {response.status_code}, 로그인 페이지에 접근할 수 없습니다.")
        return None
        
        
def download_files(session, base_url, download_url_template, output_dir):
    response = session.get(base_url)
    if response.status_code != 200:
        print(f"HTTP 에러 {response.status_code}, 페이지에 접근할 수 없습니다.")
        return

    soup = BeautifulSoup(response.text, 'html.parser')
    select_element = soup.find('select', {'name': 'no'})
    if not select_element:
        print("select 요소를 찾을 수 없습니다.")
        return

    options = select_element.find_all('option')
    if not options:
        print("option 요소를 찾을 수 없습니다.")
        return

    for option in options:
        value = option['value']
        download_url = download_url_template.format(value=value)
        download_response = session.get(download_url)   
        
        delay = 3
        retries = 3
        
        for attempt in range(retries):
            if download_response.status_code == 200:
                filename = f"{value}.txt"
                filepath = os.path.join(output_dir, filename)
                with open(filepath, 'wb') as f:
                    f.write(download_response.content)
                print(f"{filename} 다운로드 완료.")
                break
            elif response.status_code == 404:
                print(f"HTTP 에러 404, {os.path.basename(filepath)} 다운로드 실패. 재시도...")
                time.sleep(delay)  # 대기
                error.add(int(value)+1)
            else:
                print(f"HTTP 에러 {download_response.status_code}, {filename} 다운로드 실패.")
                time.sleep(delay)
                error.add(int(value)+1)
        
# 로그인
USERNAME = ''
PASSWORD = ''
session = login(USERNAME, PASSWORD)

if session:
    base_url = 'https://ncode.syosetu.com/txtdownload/top/ncode/1036066/'
    download_url_template = 'https://ncode.syosetu.com/txtdownload/dlstart/ncode/1036066/?no={value}&hankaku=0&code=utf-8&kaigyo=crlf'
    output_dir = 'downloads'
    os.makedirs(output_dir, exist_ok=True)
    download_files(session, base_url, download_url_template, output_dir)
print(error)
```
